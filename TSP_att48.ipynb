# A first overview of the implemented algorithm 

The idea is to use Ant Colony Optmisation to solve the Traveling Salesman Problem, that is to find the shortest Hamiltonian circuit in a complete graph. 

In the following notebook, we show how to use the algorithm, demonstrate its convergence and show the importance of every component of the algorithm (i.e. the pheromone and heuristic information, the numerical stability and the choice and update of the parameters and their influence on the velocity of the algorithm to converge).

```python
from utils import *
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import networkx as nx
import tsplib95
```

We choose to showcase our algorithm solving the att48 problem, which aims to find the shortest hamiltonian circuit to visit 48 capitals of the US.

```python
problem = tsplib95.load("tsp_problems/att48.tsp")
graph = problem.get_graph()
# remove self loops
graph.remove_edges_from(nx.selfloop_edges(graph))
# rename nodes to start from 0
graph = nx.relabel_nodes(graph, {node: node-1 for node in graph.nodes})
print("Nodes in graph:", graph.nodes)
print("Edges in graph:", graph.edges)
```

```python
plot_graph(graph)
```

## Collecting some statistics

The first useful statistic is the expected lenght of a hamiltonian circuit in the first iterations of our algorithm: this information is important to initialise correctly the pheromone and to have a "smooth" update of it.

```python
# compute mean distance of a path between all nodes and the distance between nodes
mean_distance = 0
n = len(graph.nodes)
for i in range(n):
    neighbours = get_neighbours(graph, i)
    mean_distance += np.mean([graph[i][n]['weight'] for n in neighbours])
    
print(mean_distance)
mean_edge = mean_distance/n
print(mean_edge)

max_v = 0
min_v = 100000
for i in range(n):
    neighbours = get_neighbours(graph, i)
    max_ = np.max([graph[i][n]['weight'] for n in neighbours])
    min_ = np.min([graph[i][n]['weight'] for n in neighbours])
    if min_ < min_v:
        min_v = min_
    if max_ > max_v:
        max_v = max_
print(min_v, max_v)
```

It is interesting, but not so unexpected, to note that a random hamiltonian circuit has a cost comparable with the average cost.

```python
# compute the distance of a random path to initialise the pheromone
random_path = list(range(len(graph.nodes)))
random.shuffle(random_path)
random_path = random_path + [random_path[0]]

cost = 0
for i in range(len(random_path)-1):
    cost += graph[random_path[i]][random_path[i+1]]['weight']
print("Cost of a random hamiltonian circuit:", cost)
```

But if we think about the first iterations of the algorithm, we know that the pheromone matrix is uniformily initialised, so the path will be mostly determined by the information coming from the heuristic, i.e. the inter-nodes distance. So we expect that in the first iterations the cost of the solutions will be close to the cost of a path determined choosing the next node as the closest node the actual node. The function to do that is defined in utils.py.

```python
min_path, cost = min_path_cost(graph)

print("Hamiltonian Circuit:", min_path)
print("Path Length:", len(min_path))
print("Heuristic min cost:", cost)
```

Another numerical problem comes in when computing the probablities to choose the next node of a path: the orders of magnitude of the pheromone matrix and the heuristic information have to be comparable, otherwise the greater of the two will overwhelm the other, loosing all the power of ACO, we will show indeed that in both cases, using only the pheromone either the heuristic information, leads to useless solutions.

So we choose to normalise the heuristic information by multiplying it by the cost of the "heuristic" path divided by the average length of an edge.

```python
norm = cost/mean_edge
print(norm)
print(min_v*norm)
print(max_v*norm)
```

We can see that the cost of the "heuristic" path stays in between the normalised min and max edge lenghts, and the normalised average length of an edge has the same lenght of the "heuristic" path.

## Description of the implemented algorithm

We already discussed initialisation of the pheromone. What we did next was to let the ants move in the graph with a probability distribution, and then update the pheromone using the well-known formula. We keep track of the best cost at each iteration. 

An improvement we brought to the algorithm is the update of the coefficients `alpha` and `beta`, the first is reduced while the second increased by a factor (0.95, 1.05) after a certain number of steps. We will see that this allows an efficient learning.

## Showing how the parameters affect the learning

First of all, a brief description of the parameters:

`alpha`: exponent of the pheromone

`beta`: exponent of the herustic information

`alpha_rate`: coefficent that multiplies alpha every `rate` iterationsù

`beta_rate`: coefficent that multiplies beta every `rate` iterations

`rho`: coefficient to regulate the dissolving of the pheromone

`max_iter`: number of iterations to perform

`k`: number of ants (i.e. paths created) at each iteration

`heuristic cost`: parameter to initialise the pheromone

`norm`: constant to normalise the weights

`opt`: wheter or not to optimise the best path at the end of the algortihm

`check_improvement`: wheter or not to check if the cost is decreasing to continue update the parameters

`update_steps`: for how many iterations to continue update alpha and beta

`elitism`: wheter or not to leverage elitism, i.e. compute the pheromone using only half of the paths (the best half)

`elitism_steps`: how many steps to wait before to start using elitism, we noticed that it is useful to leverage elitism after a warming-up phase

```python
n = len(graph.nodes)
```

We can see that trying to check the slope of the last medians and then readapt the coefficients is not that effective, while having a fixed amount of update steps is usefull. What happens in the fourth plot is that after some updates, the ratio between alpha and beta is too unbalanced, and the algorithm doesn't compute properly the probabilites to choose the next nodes when creating a circuit.

We could try different ways to balance between exploitation and exploration, but it is not easy to undeerstand why the algorithm is leveraging too much of one of the two.

A possible way to choose for how many steps perform the update of the alpha and beta is to choose a number of iterations close but less than the iteration when the drop occurs.

## Comparison with the Christofides Algorithm

We can note that the Christofides algorithm -which is deterministic- runs in much less time but returns a worse solution than than our algorithm.

Applying the local optimisation algorithm improves the solution but it is still worse.

```python
# showing that the algorithm is deterministic
tsp_sol = nx.approximation.traveling_salesman_problem(graph, cycle=True)
print(tsp_sol)
tsp_sol = nx.approximation.traveling_salesman_problem(graph, cycle=True)
print(tsp_sol)
```

```python
tsp_sol = nx.approximation.traveling_salesman_problem(graph, cycle=True)
tsp_solution = Path(tsp_sol)
cost = tsp_solution.cost
tsp_solution_opt = Path(tsp_sol)
tsp_solution_opt.opt_alg()
cost_opt = tsp_solution_opt.cost

print("Cost with Christofides:", cost)
print("Best cost with 2-opt + Crisophides:", cost_opt)
```

Note that this optimisation can be done just once, if we optimise again, even if we shuffle the indices, we get the same cost.

```python
tsp_solution_opt.opt_alg()
cost_opt = tsp_solution_opt.cost
cost_opt
```

What we can do, is to run the optimisation many times and take note of the best cost found.\

```python
history = []
best_cost = 1000000
for i in range(100):
    tsp_sol = nx.approximation.traveling_salesman_problem(graph, cycle=True)
    tsp_solution_opt = Path(tsp_sol)
    tsp_solution_opt.opt_alg()
    history.append(tsp_solution_opt.cost)
    if tsp_solution_opt.cost < best_cost:
        best_cost = tsp_solution_opt.cost

print("Best cost with 2-opt + Crisophides:", best_cost)

plt.plot(history)
plt.show()
```

## Comparison with the simple local optimisation algorithm

For curiosity, we set a random hamiltonian circuit and try to optimise it with only local optimisations.

```python
history = []
best_cost = 1000000
for i in range(100):
    random_path = list(range(len(graph.nodes)))
    random.shuffle(random_path)
    random_path = random_path + [random_path[0]]
    random_path = Path(random_path)
    random_path.opt_alg()
    if random_path.cost < best_cost:
        best_cost = random_path.cost
    history.append(random_path.cost)

print("Cost of the best optimised random path:", best_cost)
plt.plot(history)
```

We found a very nice solution with very few effort!
